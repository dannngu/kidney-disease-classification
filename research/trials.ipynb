## ensure_annotations

This is used to ensure the inputs are in the desired format to avoid problems in further steps in our fucntions returnings by data type

```python
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(x=2, y=3)
```

```python
get_prodict(x=2, y='3')
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(x=2, y='3')
```
